```python
# 이상치(Outlier) 검사
df[df['TransactionAmount'] < 0]
```

```python
# 금액이 0보다 작은 값이 있는지 확인했으나 없었으므로, 정상치 범위 확인
```

```python
Q1 = df.describe().T.loc['TransactionAmount', '25%']
Q3 = df.describe().T.loc['TransactionAmount', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

if lower_bound<0:
    lower_bound = 0

print(f'정상치 범위 : {lower_bound} ~ {upper_bound}')
```

```python
df = df[df['TransactionAmount'] <= upper_bound]
display(df)
```

그룹 분할 진행

```python
l1, l2, l3, l4 = np.percentile(new_df['Total_score'], [20,46,64,81])
print(l1,l2,l3,l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x > l1 and x <= l2:
        return 4
    if x > l2 and x <= l3:
        return 3
    if x > l3 and x <= l4:
        return 2
    return 1
    
```

```python
new_df['Level'] = new_df['Total_score'].apply(get_level)
```

```python
new_df['Level'] = new_df['Level'].replace([5,4,3,2,1],['Bronze', 'Silver','Gold','Diamond','VIP'])
new_df
```

- 지역별 분석
런던에서는 Silver등급의 회원이 가장 많이 나타났고, 뉴욕에서는 Diamond 등급의 회원이 제일 적게, Silver 등급의 회원이 가장 많이 나타났다. 파리에서도 Silver 등급의 회원이 제일 많이 나타났고, 도쿄에서도 Silver등급의 회원이 가장 많이 나타남을 알 수있다.

즉 모든 지역에서 Silver 등급의 회원이 가장 많은 것을 알 수 있었고, 파리는 다른 도시와는 달리 Bronze등급의 회원이 제일 적게 나타남을 알 수 있다. 그와 반대로, 등급이 상대적으로 높은 VIP 등급의 회원이 도쿄와 뉴욕에서 많이 있음을 알 수 있었다.

각 도시들이 대도시라는 것을 고려했을 떄, 상대적으로 VIP 등급의 회원이 많다는 것을 알 수 있다. 
- 마케팅 전략
1. B와 D 상품을 홍보 및 마케팅 강화.  
Silver 등급의 회원들이 이 상품을 가장 많이 구매하는 경향이 있으므로, 이들을 대상으로 특별한 혜택이나 할인 프로모션을 제공해 구매를 유도할 수 있고 또한 다른 등급의 회원들에게도 이 상품의 유용성을 강조하는 캠페인을 통해 다른 등급의 회원들도 홍보하여 구매를 촉진할 수 있습니다.  

2. C 상품의 프로모션을 VIP 및 Diamond 등급 회원에게 집중.  
C 상품은 VIP 및 Diamond 등급의 회원들이 주로 많이 사는 경향이 보이기에 이에 따라 이 등급의 회원들에게 혜택을 제공하여 C 상품의 매출을 더 높이는 전략이 효과적일 수 있습니다.

3. Gold 등급의 회원들을 대상으로 다양한 상품 소개 및 이벤트 진행  
Gold 등급의 회원들은 모든 제품을 균일하게 구매하는 경향이 있기에, 이 등급의 회원들에게는 더 A,B,C,D를 제외한 다른 제품 소개 및 할인 프로모션을 진행한다면, 구매 다양성을 더욱 높일 수 있을것으로 보입니다.

4. 각 지역의 회원 등급분포를 고려  
파리에서는 Silver 등급 회원이 가장 많기에, Silver 등급 회원을 위한 마케팅 자원을 집중할 수 있을 것으로 판단됩니다.  
또한, 뉴욕에서는 Diamond 등급 회원이 다른 등급의 회원보다 적은 수이기에, 이 회원의 수를 유치하기 위한 캠페인을 진행하여 고객 이탈을 방지할 뿐만 아니라 이 등급의 회원의 수를 늘릴 수 있을 것입니다.  
런던과 도쿄는, 다른 등급에 비해 약소하게 Silver 회원의 수가 더 많지만, 모든 회원을 대상으로 진행할 수 있는 프로모션을 짜는 것이 효과적일 것입니다.
추가적으로 도쿄에서는 다른 지역에 비해 Diamond와 VIP 회원의 수가 더 많기에 이들을 대상으로 한 독점적인 이벤트나 혜택을 제공하여 매출을 증대할 수 있습니다.
