```python
# 가져온 값을 설정
hap_gdp_df = hap_df.T.loc[['happiness_score','gdp_per_capita']].T
display(hap_gdp_df)
```

```python
#데이터 변환
import numpy as np
print(hap_gdp_df.columns)
print(hap_gdp_df.index)
hap_gdp_df.info()
hap_gdp_df.dtypes
hap_gdp_df = hap_gdp_df.astype(np.float32)
hap_gdp_df.dtypes

display(hap_gdp_df)
```

```python
hap_gdp_df.describe()
```

```python
# 그래프로 표기
import matplotlib.pyplot as plt

hap_gdp_df.plot(x='gdp_per_capita', y='happiness_score', fontsize='4', kind='scatter', grid=True, ylabel='happiness_score')
```

```python
display(hap_gdp_df.describe().T)
```

```python
# 정상치 범위 확인
from scipy.stats import iqr

iqr_value = iqr(hap_gdp_df.happiness_score)
print(iqr_value)

Q1 = hap_gdp_df.describe().T.loc['happiness_score', '25%']
Q3 = hap_gdp_df.describe().T.loc['happiness_score', '75%']

print(f'Q1 값 : {Q1}, Q3 값 : {Q3}')
lower_bound = Q1 - 1.5*iqr_value
upper_bound = Q3 + 1.5*iqr_value
print(f'정상치 범위 : {lower_bound}~{upper_bound}')
```

```python
# 정상치 범위 벗어난 값 확인
display(hap_gdp_df[hap_gdp_df['happiness_score'] < lower_bound])
display(hap_gdp_df[hap_gdp_df['happiness_score'] > upper_bound])
```

```python
# 정상치 벗어난 값 제거
display(hap_gdp_df)
hap_gdp_df = hap_gdp_df.drop(index=[136], axis=0)
display(hap_gdp_df)
```

```python
#기초 통계량 확인
hap_gdp_df.describe()
```

```python
display(hap_gdp_df.describe().T)
hap_gdp_df.describe().T.loc['gdp_per_capita', '25%']
```

```python
# 그룹화를 위해 index 추가
display(hap_gdp_df.tail(5))
hap_gdp_df[hap_gdp_df['gdp_per_capita']< hap_gdp_df.describe().T.loc['gdp_per_capita', '25%']]
```

```python
# 그룹화를 위해 25, 50, 75퍼센트 별 구간 지정
```

```python
hap_gdp_df.loc[hap_gdp_df['gdp_per_capita'] < hap_gdp_df['gdp_per_capita'].describe().loc['25%'], 'income'] = 'Low income'

hap_gdp_df.loc[(hap_gdp_df['gdp_per_capita'].describe().loc['25%'] <= hap_gdp_df['gdp_per_capita']) & \
(hap_gdp_df['gdp_per_capita'] < hap_gdp_df['gdp_per_capita'].describe().loc['50%']), 'income'] = 'Lower middle income'

hap_gdp_df.loc[(hap_gdp_df['gdp_per_capita'].describe().loc['50%'] <= hap_gdp_df['gdp_per_capita']) & \
(hap_gdp_df['gdp_per_capita'] < hap_gdp_df['gdp_per_capita'].describe().loc['75%']), 'income'] = 'Upper middle income'

hap_gdp_df.loc[hap_gdp_df['gdp_per_capita'] >=  hap_gdp_df['gdp_per_capita'].describe().loc['75%'], 'income'] = 'High income'
display(hap_gdp_df)
```

```python
# 그룹화 한 후 데이터 확인
hp_mean_df = hap_gdp_df.groupby('income').mean('happiness_score').reset_index()
display(hp_mean_df)
hp_min_df = hap_gdp_df.groupby('income').min('happiness_score').reset_index()
display(hp_min_df)
hp_max_df = hap_gdp_df.groupby('income').max('happiness_score').reset_index()
display(hp_max_df)
hp_std_df = hap_gdp_df.groupby('income')['happiness_score'].std().reset_index()
display(hp_std_df)
```

```python
# 그래프로 확인
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

hp_mean_df.plot(title='mean', x='income', y='happiness_score', fontsize='4', kind='line', grid=True, ax=axes[0], ylabel="score")
hp_min_df.plot(title='min', x='income', y='happiness_score', fontsize='4', kind='line', grid=True, ax=axes[1], ylabel="score")
hp_max_df.plot(title='max', x='income', y='happiness_score', fontsize='4', kind='line', grid=True, ax=axes[2], ylabel="score")
hp_std_df.plot(title='std', x='income', y='happiness_score', fontsize='4', kind='line', grid=True, ax=axes[3], ylabel="score")
```

```python
# 전체 그래프로 확안
hap_gdp_df.plot(x='gdp_per_capita', y='happiness_score', fontsize='4', kind='scatter', grid=True, ylabel='happiness_score')
```

```python
# 정상치 범위 벗어난 값 제거
# 위에서 happiness_score의 정상치 범위를 벗어난 값은 136번의 값인 것을 확인했기 때문에, 그 값을 제거

hap1_df = hap1_df.drop(index=[136], axis=0)
display(hap1_df)
```

```python
# gdp와 마찬가지로 social_support 구간을 구분
hap1_df.loc[hap1_df['social_support'] < hap1_df['social_support'].describe().loc['25%'], 'support'] = 'Low support'

hap1_df.loc[(hap1_df['social_support'].describe().loc['25%'] <= hap1_df['social_support']) & \
(hap1_df['social_support'] < hap1_df['social_support'].describe().loc['50%']), 'support'] = 'Lower middle support'

hap1_df.loc[(hap1_df['social_support'].describe().loc['50%'] <= hap1_df['social_support']) & \
(hap1_df['social_support'] < hap1_df['social_support'].describe().loc['75%']), 'support'] = 'Upper middle support'

hap1_df.loc[hap1_df['social_support'] >=  hap1_df['social_support'].describe().loc['75%'], 'support'] = 'High support'
display(hap1_df)
```

```python
# 그룹화 한 후 데이터 확인
hp_mean_df = hap1_df.groupby('support').mean('happiness_score').reset_index()
display(hp_mean_df)
hp_min_df = hap1_df.groupby('support').min('happiness_score').reset_index()
display(hp_min_df)
hp_max_df = hap1_df.groupby('support').max('happiness_score').reset_index()
display(hp_max_df)
hp_std_df = hap1_df.groupby('support')['happiness_score'].std().reset_index()
display(hp_std_df)
```

### 종합분석

- happiness_score(이하 행복지수)와 각 컬럼명간의 연관성을 비교해보면 다음과 같다.
- gdp_per_capita(1인당 GDP)와  social_support(사회적 지지를 측정하는 척도), healthy_life_expectancy(건강한 기대 수명)는  
  전체적으로 우상향 그래프의 모양을 나타내고, 각 구간별 표준편차의 값도 다른 컬럼에 비해 작은편에 속한다. 
- freedom_to_make_life_choices(삶의 선택에 있어 자유로움의 척도)는 우상향 그래프의 형태를 보이지만,  
  각 구간별 표준편차 값이 상대적으로 크게 나타난다.
- generosity(관대함을 반영하는 척도), perception_of_corruption(국가 내 부패에 대한 인식의 척도)는  
  전체적인 그래프 모양을 판단하기 어려울 뿐만 아니라 각 구간별 표준편차의 값도 상대적으로 크게 나타난다.

- 표준편차의 값이 상대적으로 크게 나타난다는 것은 그 데이터들이 평균값에서 분산되어 있다는 것을 의미한다.  
  결국 1인당 GDP와 사회적 지지를 측정하는 척도, 건강한 기대 수명 항목이 전체적으로 우상향의 그래프모양을 나타내면서  
  동시에 표준편차의 값이 상대적으로 작게 나타나고 있기 때문에 행복지수와 가장 큰 연관성이 있다고 판단할 수 있다.

- 이를 종합해보면, 행복지수에 가장 큰 영향을 미치는 요인은 경제적 안정과 사회적 지지, 그리고 건강에 관련된 요소라고 볼 수 있다.  
  이러한 경향은 서부유럽과 같은 발전된 지역에서 뚜렷하게 나타나며,  이 지역들은 평균적으로 높은 행복지수를 가지고 있으며 행복지수의 변동이 작다.  
  반면에 중동아프리카와 북아프리카와 같은 지역은 행복지수가 낮으며, 이 지역들은 행복지수의 변동이 크고 다양한 경향을 가지고 있다고 볼 수 있다.
