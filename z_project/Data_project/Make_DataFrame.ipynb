```python
def get_level(x):
    if x == 9:
        return 'Royal Flush'
    elif x == 8:
        return 'Straight Flush'
    elif x == 7:
        return 'Four Card'
    elif x == 6:
        return 'Full House'
    elif x == 5:
        return 'Flush'
    elif x == 4:
        return 'Straight'
    elif x == 3:
        return 'Three of a kind'
    elif x == 2:
        return 'Two pair'
    elif x == 1:
        return 'One pair'
    else:
        return 'Top'
```

```python
import pandas as pd
text = pd.read_table('./datasets/File 190.txt', header=None, names=['des'])
# 주어진 텍스트
display(text.iloc[0])
display(text.iloc[54])
text
```

```python
game_started_indices = text[text['des'].str.contains('Game started')].index
game_ended_indices = text[text['des'].str.contains('Game ended')].index


game_data = []

for start_idx, end_idx in zip(game_started_indices, game_ended_indices):
    game_text = text.iloc[start_idx:end_idx+1]
    game_data.append(game_text)
print(len(game_data))
display(game_data[8])
```

각각의 데이터프레임 생성


```python
import re
```

```python
import torch, gc
gc.collect()
torch.cuda.empty_cache()

        
import warnings
warnings.filterwarnings(action='ignore')
```

```python
dfs = {}
player_card_pattern = r'\[([^\[\]]+)\]'

for i in range(len(game_data)):
    dfs[i] = pd.DataFrame(columns=['GameNo', 'SeatNo','Button','SmallBlind','PlayerName','SeedMoney', \
                                  'PlayerCard', 'raise', 'fold', 'bets', 'call','check', \
                                   'FLOP','TURN','RIVER','Board','Win'])
    game_no = i
    g_df = game_data[i]

    for line in g_df['des']:
        if 'Seat' in line and 'button' not in line and ':' in line and '(' in line and 'Collects:' not in line:
            seat_no = line.split(':')[0].split(' ')[-1]
            player_name = line.split(' ')[2]
            player_money = line.split(' ')[-1].replace(').', '').replace('(', '').strip()
            #player_money = re.sub(r'[()\!]', ' ', player_money).strip().split(' ')[0]
            player_money = float(player_money)
            dfs[i] = pd.concat([dfs[i], pd.DataFrame({'GameNo': [game_no], 'SeatNo': [seat_no], \
                                          'PlayerName': [player_name] ,'SeedMoney': [player_money]})], ignore_index=True)
            
            continue
            
    dfs[i]['Button'] = dfs[i]['Button'].fillna(0)
    dfs[i]['raise'] = dfs[i]['raise'].fillna(0)
    dfs[i]['fold'] = dfs[i]['fold'].fillna(0)
    dfs[i]['bets'] = dfs[i]['bets'].fillna(0)
    dfs[i]['call'] = dfs[i]['call'].fillna(0)
    dfs[i]['check'] = dfs[i]['check'].fillna(0)
    dfs[i]['Win'] = dfs[i]['Win'].fillna(0)
    dfs[i]['SmallBlind'] = dfs[i]['SmallBlind'].fillna(0)
    
    player_cards= {}        
    for line in g_df['des']:            
        if 'is the button' in line:
            seat_no = int(line.split(' ')[1])
            #print(seat_no)
            dfs[i].loc[dfs[i].SeatNo==seat_no, 'Button'] = 1

        if 'received card' in line:
            #print(line)
            card = (line.split(':')[-1].split('[')[1].split(']')[0])
            #print(card)
            player_name = line.split(' ')[1]
            #print(player_name)
            if player_name not in player_cards:
                player_cards[player_name] = []
            
            player_cards[player_name].append(card)

        if 'folds' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']==player_name, 'fold'] = 1
    
        if 'raises' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']==player_name, 'raise'] += 1

        if '*' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']==player_name, 'Win'] = 1

        if 'FLOP' in line:
            board_card = line.split(':')[-1].replace('[', '').replace(']', '').strip()
            #print(board_card)
            dfs[i].loc[:,'FLOP'] = board_card

        if 'RIVER' in line:
            board_card = line.split(':')[-1].split(' ')[-1].replace('[', '').replace(']', '').strip()
            dfs[i].loc[:,'RIVER'] = board_card
        
        if 'TURN' in line:
            board_card = line.split(':')[-1].split(' ')[-1].replace('[', '').replace(']', '').strip()
            dfs[i].loc[:,'TURN'] = board_card

        if 'Board:' in line:
            board_card = line.split(':')[-1].replace('[', '').replace(']', '').strip()
            dfs[i].loc[:, 'Board'] = board_card

        if 'shows:' in line:
            player_name = line.split(' ')[1].strip()
            match = re.search(player_card_pattern, line)
            if match:
                player_card = match.group(1)
                dfs[i].loc[dfs[i]['PlayerName'] == player_name, 'PlayerCard'] = ''.join(player_card)

        if 'bets' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']== player_name, 'bets'] += 1

        if 'calls' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']== player_name, 'call'] += 1

        if 'checks' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']== player_name, 'check'] += 1

        if 'has small blind' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']== player_name, 'bets'] += 1
            dfs[i].loc[dfs[i]['PlayerName']== player_name, 'SmallBlind'] = 1
            
        if 'has big blind' in line:
            player_name = line.split(' ')[1].strip()
            dfs[i].loc[dfs[i]['PlayerName']== player_name, 'bets'] += 1
            
    for player_name, cards in player_cards.items():
        dfs[i].loc[dfs[i]['PlayerName'] == player_name, 'PlayerCard'] = ' '.join(cards)
```
