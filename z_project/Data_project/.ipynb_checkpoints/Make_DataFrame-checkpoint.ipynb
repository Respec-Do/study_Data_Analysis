```python
def get_Suit(x):
    if x == 1:
        return 'Hearts'
    elif x == 2:
        return 'Spades'
    elif x == 3:
        return 'Diamonds'
    else:
        return 'Clubs'
```

```python
def get_level(x):
    if x == 9:
        return 'Royal Flush'
    elif x == 8:
        return 'Straight Flush'
    elif x == 7:
        return 'Four Card'
    elif x == 6:
        return 'Full House'
    elif x == 5:
        return 'Flush'
    elif x == 4:
        return 'Straight'
    elif x == 3:
        return 'Three of a kind'
    elif x == 2:
        return 'Two pair'
    elif x == 1:
        return 'One pair'
    else:
        return 'Top'
```

```python
df['Level'] = df['Class'].apply(get_level)
df
```

```python
import pandas as pd

# 데이터 추출

df = pd.DataFrame(columns=['GameNo', 'SeatNo','PlayerName','SeedMoney', 'PlayerCard', 'raise', 'fold', 'bets', 'Button','Win'])
game_no = 1
game_started = False
player_lines = text.strip().split('\n')

for line in player_lines:
    print(line)
    if 'Game started' in line:
        game_started = True
    if game_started and 'Seat' in line and 'button' not in line:
        seat_no = int(line.split(':')[0].split(' ')[1])
        player_name = (line.split(' ')[2].strip())
        player_money = (line.split(' ')[3])
        player_money = re.sub(r'[()\!]', ' ', player_money).strip()
        player_money = float(player_money.split(' ')[0])
        
        df = pd.concat([df, pd.DataFrame({'GameNo': [game_no], 'SeatNo': [seat_no], \
                                          'PlayerName': [player_name] ,'SeedMoney': [player_money]})], ignore_index=True)

    elif 'Game ended' in line:
        game_started = False
        game_no += 1

player_cards = {}

for line in player_lines:
    
    if 'is the button' in line:
        seat_no = int(line.split(' ')[1])
        df.loc[df['SeatNo']==seat_no, 'Button'] = 1

    if '[' in line:
        card = (line.split(':')[1].split('[')[1].split(']')[0])
        player_name = line.split(' ')[1]
        print(player_name)
        if player_name not in player_cards:
            player_cards[player_name] = []
        
        player_cards[player_name].append(card)

    if 'folds' in line:
        player_name = line.split(' ')[1]
        df.loc[df['PlayerName']==player_name, 'fold'] = True

    if 'mucks' in line:
        player_name = line.split(' ')[1]
        df.loc[df['PlayerName']==player_name, 'Win'] = 1

    if 'raises' in line:
        player_name = line.split(' ')[1]
        df.loc[df['PlayerName']==player_name, 'raise'] = 1
print(player_cards.items())
for player_name, cards in player_cards.items():
    df.loc[df['PlayerName'] == player_name, 'PlayerCard'] = ' ,'.join(cards)
    







display(df)

```
